```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import t

# Sample sizes
n = np.arange(0, 251)

# Confidence levels
conf_levels = [0.99, 0.95, 0.90]


plt.figure(figsize=(10, 6))


for conf in conf_levels:
    alpha = 1 - conf
    t_crit = t.ppf(1 - alpha / 2, df=n - 2)
    r_crit = np.sqrt(t_crit**2 / (t_crit**2 + (n - 2)))
    plt.plot(n, r_crit, label=f'{int(conf * 100)}% Confidence')


plt.xlabel('Sample Size')
plt.ylabel('Critical Value of Pearson Correlation')
plt.title('Critical Values of Pearson Correlation vs. Sample Size')
plt.legend()

plt.show()
```
